# VQAv2 — Quick Exploration
Verify dataset loading, inspect sample images, check answer distribution.

```python
import sys; sys.path.insert(0, '..')
from src.dataset import load_vqa_subset
train_ds, eval_ds = load_vqa_subset(sample_size=200, seed=42)
print(train_ds, eval_ds)
```

```python
import matplotlib.pyplot as plt
fig, axes = plt.subplots(2, 3, figsize=(12, 7))
for ax, ex in zip(axes.flat, train_ds.select(range(6))):
    ax.imshow(ex['image'])
    ax.set_title(f"Q: {ex['question']}\nA: {ex['answers'][0]['answer']}", fontsize=8)
    ax.axis('off')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
qtypes = pd.Series([ex['question_type'] for ex in train_ds])
qtypes.value_counts().head(10).plot(kind='barh', title='Question types')
```
